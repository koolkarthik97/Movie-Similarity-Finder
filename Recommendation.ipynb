```python
import numpy as np
import pandas as pd
```

### Preprocess ratings.dat and movies.dat file

```python
column_names = ['user_id' , 'item_id' , 'rating' , 'timestamp']

with open('ratings.dat', 'r') as fp:
    lines = fp.readlines()

    
```

```python
new_file=open('processed_ratings.dat','w')
for line in lines:
    line=line.replace("::"," ")
    new_file.write(line)
```

```python
new_file.close()
```

```python
dataframe = pd.read_csv('processed_ratings.dat',sep=' ',names=column_names)
dataframe.head()
```

```python
with open('movies.dat', 'r',encoding="ISO-8859-1") as fp:
    movies = fp.readlines()
```

```python
movie_file = open('processed_movies.dat','w')
for movie in movies:
    split_movie = movie.split("::")
    write_str = split_movie[0]+"\t"+split_movie[1]
    movie_file.write(write_str+"\n")
movie_file.close()
    
```

```python
column_names = ['item_id' , 'movie_title']
movie_titles = pd.read_csv("processed_movies.dat",sep = '\t',names = column_names)
movie_titles.head()
```

### Compute correlations 

```python
df = pd.merge(dataframe,movie_titles,on='item_id')
```

```python
df.head()
```

```python
movie_matrix = df.pivot_table(index='user_id',columns='movie_title',values='rating')
```

```python
ratings = pd.DataFrame(df.groupby('movie_title')['rating'].mean())
ratings['number of ratings'] = pd.DataFrame(df.groupby('movie_title')['rating'].count())
```

```python
ratings.sort_values('number of ratings',ascending=False).head(10)
```

```python
movie_matrix.head()
```

```python
american_beauty_user_ratings = movie_matrix["American Beauty (1999)"]
silence_of_the_lambs_user_ratings = movie_matrix["Silence of the Lambs, The (1991)"]
```

```python
similar_to_american_beauty = movie_matrix.corrwith(american_beauty_user_ratings)
similar_to_silence_of_the_lambs = movie_matrix.corrwith(silence_of_the_lambs_user_ratings)
```

```python
corr_american_beauty = pd.DataFrame(similar_to_american_beauty,columns=['Correlation'])
corr_silence_of_the_lambs = pd.DataFrame(similar_to_silence_of_the_lambs,columns=['Correlation'])
```

```python
corr_american_beauty.dropna(inplace=True)
corr_silence_of_the_lambs.dropna(inplace=True)
```

```python
corr_american_beauty.head()
```

```python
ratings.head()
```

```python
corr_silence_of_the_lambs=corr_silence_of_the_lambs.join(ratings['number of ratings'], on='movie_title')
```

```python
corr_american_beauty=corr_american_beauty.join(ratings['number of ratings'], on='movie_title')
```

```python
corr_american_beauty[corr_american_beauty['number of ratings']>50].sort_values('Correlation',ascending=False).head()
```
